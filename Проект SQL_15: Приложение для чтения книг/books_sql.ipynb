## Посчитайте, сколько книг вышло после 1 января 2000 года

## Для каждой книги посчитайте количество обзоров и среднюю оценку

```python
query = '''
SELECT books.title AS title,
       books.book_id AS book_id, 
       COUNT(DISTINCT reviews.review_id) AS reviews_cnt, 
       AVG(ratings.rating) AS avg_rating
FROM books
LEFT JOIN ratings ON ratings.book_id = books.book_id
LEFT JOIN reviews ON reviews.book_id = books.book_id
GROUP BY books.title, books.book_id
ORDER BY reviews_cnt DESC
LIMIT 5
'''
books_sql(query)
```

Больше всего обзоров на книгу "Сумерки", хотя рейтинг 3.66 баллов.

## Определите издательство, которое выпустило наибольшее число книг толще 50 страниц — так вы исключите из анализа брошюры

```python
query = '''
SELECT publ.publisher as publisher_name,
       COUNT(DISTINCT books.book_id) as books_cnt
FROM publishers as publ
LEFT JOIN books on publ.publisher_id = books.publisher_id
WHERE books.num_pages > 50
GROUP BY publisher_name
ORDER BY books_cnt DESC
LIMIT 5;
'''
books_sql(query)
```

В этом рейтинге на первом месте издательство Penguin Books, которое выпустило 42 книги толще 50 страниц.

## Определите автора с самой высокой средней оценкой книг — учитывайте только книги с 50 и более оценками

```python
query = '''
SELECT author_rating.author AS author,  
       AVG(author_rating.avg_rating) AS rating 
FROM (
SELECT authors.author AS author,
       books.book_id AS id, 
       AVG(ratings.rating) AS avg_rating,
       COUNT(ratings.rating) AS rating_cnt
       FROM authors
       INNER JOIN books ON books.author_id = authors.author_id
       INNER JOIN ratings ON ratings.book_id = books.book_id
       GROUP BY author, id
       HAVING COUNT(ratings.rating) > 50) AS author_rating
GROUP BY author
ORDER BY rating DESC
LIMIT 5
'''

books_sql(query)
```

J.K. Rowling/Mary GrandPré - автор с самой высокой средней оценкой книг.

## Посчитайте среднее количество обзоров от пользователей, которые поставили больше 48 оценок

```python
query = '''
WITH avg_user_reviews as
    (SELECT username,
     COUNT(review_id) as review_cnt
     FROM reviews
     WHERE username in (
     SELECT username
     FROM ratings
     GROUP BY username
     HAVING COUNT(rating) > 48)
     GROUP BY username)
SELECT AVG(review_cnt)
FROM avg_user_reviews    
'''

books_sql(query)
```
